# 读取路径里的数据loss或mAP

```python
def get_loss_list(loss_path):
    loss_values = []
    with open(loss_path, 'r', encoding='utf-8') as file:
        for line in file:
            # 将每行的损失值转换为浮点数并添加到列表中
            loss_values.append(float(line.strip()))

    return loss_values

def get_mAP_list(mAP_path):
    mAP_values = []
    with open(mAP_path, 'r', encoding='utf-8') as file:
        for line in file:
            # 将每行的损失值转换为浮点数并添加到列表中
            mAP_values.append(float(line.strip()))

    return mAP_values

# loss_path = "../RecordTrain/CNNF/loss_DPSH_CNNF_cifar10-1_12.txt"
# loss_list = get_loss_list(loss_path)
# print(loss_list)
# print(len(loss_list))
# mAP_path = "../RecordTrain/CNNF/mAP_DPSH_CNNF_cifar10-1_12.txt"
# mAP_list = get_loss_list(mAP_path)
# print(mAP_list)
# print(len(mAP_list))
```

# 绘制LOSS和mAP的曲线函数

```python
def draw_loss(path_list, colors, labels, title):
    # 获取四个损失值列表
    loss_lists = [get_loss_list(path) for path in path_list]

    # 确定 x 轴数据，假设所有损失值列表的长度都相同
    x = list(range(1, len(loss_lists[0]) + 1))

    # 绘制折线图
    plt.figure(figsize=(10, 6))

    for i in range(len(loss_lists)):
        plt.plot(x, loss_lists[i], marker='o', linestyle='-', color=colors[i], label=labels[i])

    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # 显示图形
    plt.show()
```

```python
def draw_mAP(path_list, colors, labels, title):
    # 获取四个损失值列表
    mAP_lists = [get_mAP_list(path) for path in path_list]

    # 确定 x 轴数据，假设所有损失值列表的长度都相同
    x = list(range(1, len(mAP_lists[0]) + 1))

    # 绘制折线图
    plt.figure(figsize=(10, 6))

    for i in range(len(mAP_lists)):
        plt.plot(x, mAP_lists[i], marker='o', linestyle='-', color=colors[i], label=labels[i])

    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('mAP')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # 显示图形
    plt.show()
```
